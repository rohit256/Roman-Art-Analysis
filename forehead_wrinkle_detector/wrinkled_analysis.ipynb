```python
import cv2
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)

import matplotlib.pyplot as plt

image = cv2.imread('/home/rohit/Desktop/Roman-Art-Analysis/forehead_wrinkle_detector/wrinkled/a14.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.GaussianBlur(image,(5,5),0)

plt.imshow(image)
plt.show()
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(image_gray,  cmap='gray')
plt.show()
```

```python
sobely = cv2.Laplacian(image_gray, cv2.CV_8UC1)

plt.imshow(sobely,  cmap='gray')
plt.show()
```

```python
height, width= sobely.shape
print(height,width)
print(sobely)
for i in range(height):
    for j in range(width):
        sobely
param_1 = cv2.sumElems(sobely)[0]/(height * width)
print(param_1)
```
